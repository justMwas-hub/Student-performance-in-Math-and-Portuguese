```python
def litmus(grade):
    if grade >= 10:
        return 'pass'
    else:
        return 'fail'
```

```python
def level_five(grade):
    if grade < 10:
        return 'V'
    elif grade >= 10 and grade < 12:
        return 'IV'
    elif grade >= 11 and grade < 14:
        return 'III'
    elif grade >= 13 and grade < 16:
        return 'II'
    else:
        return 'I'
```

```python
df["Erasmus_gr"] = df["G3"].apply(level_five)
```

```python
df
```

```python
plt.figure(figsize=(6,4), dpi=150)
sns.countplot(data=df, x="Erasmus_gr", order=df['Erasmus_gr'].value_counts().index)
```

```python
plt.figure(figsize=(6,4), dpi=150)
sns.countplot(data=df, x="results")
```

```python
df.to_csv("school_mat_performance_Erasmus.csv", index_label=False)
```
